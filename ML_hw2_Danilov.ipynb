```python
import torch
import torchvision
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
from mpl_toolkits.axes_grid1 import ImageGrid
```

```python
def draw(img):
    fig = plt.figure(figsize=(20, 20))
    grid = ImageGrid(fig, 111, nrows_ncols=(1, 3), axes_pad=0.02)
    for ax, im in zip(grid, img):
        ax.imshow(im)
    plt.show()
```

```python
pic = Image.open('pic.jpg').convert('RGB')
input_data = np.array(pic)
input_data = input_data.transpose((2, 0, 1))
```

```python
conv1 = torch.nn.Conv2d(in_channels=3, out_channels=1, kernel_size=(3, 3), stride=1)
relu = torch.nn.ReLU()
SiLU= torch.nn.SiLU()
Sigmoid = torch.nn.Sigmoid()
Tanh = torch.nn.Tanh()
```

```python
input_tensor = torch.from_numpy(input_data).to(torch.float32)
output_tensor = conv1(input_tensor)
np_out = output_tensor.detach().numpy()
draw(np_out)
```

```python
output_tensor_relu = relu(output_tensor)
np_out_relu = output_tensor_relu.detach().numpy()
draw(np_out_relu)
```

```python
output_tensor_silu = SiLU(output_tensor)
np_out_silu = output_tensor_silu.detach().numpy()
draw(np_out_silu)
```

```python
output_tensor_sigmoid = Sigmoid(output_tensor)
np_out_sigmoid = output_tensor_sigmoid.detach().numpy()
draw(np_out_sigmoid)
```

```python
output_tensor_tanh = Tanh(output_tensor)
np_out_tanh = output_tensor_tanh.detach().numpy()
draw(np_out_tanh)
```
