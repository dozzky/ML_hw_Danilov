```python
import torch
import torchvision
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
pic = Image.open('pic.jpg').convert('RGB')

plt.figure(figsize=(15, 15))
plt.subplot(2, 1, 1)
plt.imshow(pic)

pic = torchvision.transforms.ToTensor()(pic)
pic = pic.unsqueeze(0)
conv = torch.nn.Conv2d(3, 3, 3)
pic = conv(pic)
pic = pic.squeeze(0)
pic = torchvision.transforms.ToPILImage()(pic)

plt.subplot(2, 1, 2)
plt.imshow(pic)
```
